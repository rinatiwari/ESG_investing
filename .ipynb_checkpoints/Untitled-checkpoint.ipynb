```python
import pandas as pd

data_csv = pd.read_csv("Industry_grouping1.csv")

```

```python
data_csv = data_csv.reset_index(drop=True)
data_csv = data_csv.round(2)
data_csv.head()
```

```python
industry_top_10 = data_csv.sort_values(by=['ESG Risk Score'], ascending=True)
industry_top_10 = industry_top_10.head(10)
industry_top_10 = industry_top_10.reset_index(drop=True)
industry_top_10
```

```python
industry_top_10_Environmental = data_csv.sort_values(by=['Environment Risk Score'], ascending=True)
industry_top_10_Environmental = industry_top_10_Environmental.head(10)
industry_top_10_Environmental = industry_top_10_Environmental.reset_index(drop=True)
industry_top_10_Environmental
```

```python
industry_top_10_Social = data_csv.sort_values(by=['Social Risk Score'], ascending=True)
industry_top_10_Social = industry_top_10_Social.head(10)
industry_top_10_Social = industry_top_10_Social.reset_index(drop=True)
industry_top_10_Social
```

```python
industry_top_10_Governance = data_csv.sort_values(by=['Governance Risk Score'], ascending=True)
industry_top_10_Governance = industry_top_10_Governance.head(10)
industry_top_10_Governance = industry_top_10_Governance.reset_index(drop=True)
industry_top_10_Governance
```

```python
data_csv.to_json("ESG_Industry_Table.json")
industry_top_10.to_json("ESG_Industry10_Table.json")
industry_top_10_Environmental.to_json("E_Industry10_Table.json")
industry_top_10_Social.to_json("S_Industry10_Table.json")
industry_top_10_Governance.to_json("G_Industry10_Table.json")
```
